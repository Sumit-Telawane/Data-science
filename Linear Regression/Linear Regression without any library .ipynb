```python
# Given data
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,5,4,7,8,10,11,6.8]
```

```python
def linear_regression_coefficients(x, y):
    # Calculate the mean of x and y
    mean_x = sum(x) / len(x)
    mean_y = sum(y) / len(y)

    # Initialize numerator and denominator
    numerator = 0
    denominator = 0

    # Calculate the numerator and denominator separately using a for loop
    for xi, yi in zip(x, y):
        numerator += (xi - mean_x) * (yi - mean_y)
        denominator += (xi - mean_x)**2

    # Calculate the slope (m)
    m = numerator / denominator

    # Calculate the y-intercept (c)
    c = mean_y - m * mean_x

    return m, c

# Example usage
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 5, 4, 7, 8, 10, 11, 6.8]

slope, intercept = linear_regression_coefficients(x, y)
print(f"Slope (m): {slope}")
print(f"Y-intercept (c): {intercept}")
```

```python
# Plot the scatter plot
plt.scatter(x, y, label='Data Points')

# Plot the linear regression line using a for loop
regression_line = []
for xi in x:
    regression_line.append(slope * xi + intercept)

plt.plot(x, regression_line, color='red', label='Linear Regression Line')

# Add labels and a legend
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title("Linear Regression")

# Show the plot
plt.show()
```
