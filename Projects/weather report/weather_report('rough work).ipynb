```python
import datetime as dt
import requests

# Base URL for the OpenWeatherMap API
BASE_URL = "http://api.openweathermap.org/data/2.5/weather?"
# API key for authentication
API_KEY = 'ea642ff5d0165717d77c25f8bdb605da'

# Function to convert temperature from Kelvin to Celsius
def kelvin_to_celsius(kelvin):
    celsius = kelvin - 273.15
    return celsius

# City for which to fetch weather data
districts = [
    "Mumbai", "Pune", "Nagpur", "Aurangabad", "Nashik", "Amravati", "Kolhapur",
    "Thane", "Jalgaon", "Satara", "Solapur", "Bhandara", "Wardha",
    "Chandrapur", "Akola", "Washim", "Buldhana", "Hingoli",
    "Parbhani", "Jalna", "Osmanabad", "Latur", "Beed", "Yavatmal",
    "Ratnagiri", "Palghar", "Nandurbar", "Dhule", "Jalgaon",
    "Ahmednagar", "Sangli", "Sangli", "Kolhapur", "Solapur", "Satara"
]
for city in districts:
    
    # Construct the complete API URL
    url = BASE_URL + "appid=" + API_KEY + "&q=" + city
    # Send a GET request to the API and get the response in JSON format
    response = requests.get(url).json()
    
    
    # Extract temperature in Kelvin from the response
    temp_kelvin = response['main']['temp']
    # Convert temperature to Celsius
    temp_celsius = kelvin_to_celsius(temp_kelvin)
    
    # Extract 'feels like' temperature in Kelvin from the response
    feels_like_kelvin = response['main']['feels_like']
    # Convert 'feels like' temperature to Celsius
    feels_like_celsius = kelvin_to_celsius(feels_like_kelvin)
    
    # Extract humidity from the response
    humidity = response['main']['humidity']
    
    # Extract weather description from the response
    description = response['weather'][0]['description']
    
    # Convert sunrise time from timestamp to datetime and adjust for timezone
    sunrise_time = dt.datetime.utcfromtimestamp(response['sys']['sunrise'] + response['timezone'])
    # Convert sunset time from timestamp to datetime and adjust for timezone
    sunset_time = dt.datetime.utcfromtimestamp(response['sys']['sunset'] + response['timezone'])
    
    # Extract wind speed from the response
    wind_speed = response['wind']['speed']
    
    # Create a dictionary with all the weather data
    weather_data = {
        'city': city,
        'temperature': f"{temp_celsius:.2f}°C",
        'feels_like_celsius': f"{feels_like_celsius:.2f}°C",
        'humidity': f'{humidity}%',
        'wind_speed': f'{wind_speed}m/s',
        'sunrise_time': sunrise_time.strftime('%Y-%m-%d %H:%M:%S'),
        'sunset_time': sunset_time.strftime('%Y-%m-%d %H:%M:%S'),
        'description': description,
    }
    #store_weather_data(weather_data)
    # Print or return the weather data dictionary
    print(weather_data)
```
