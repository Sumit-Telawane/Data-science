# Exploratory Data Analysis(EDA)

# Definition

Exploratory Data Analysis (EDA) is an approach to analyzing datasets to summarize their main characteristics, often using visual methods and statistical techniques. The primary goal of EDA is to gain insights, uncover patterns, identify anomalies, and understand the underlying structure of the data before conducting more formal statistical analyses or modeling. EDA involves examining the data to describe its properties, distributions, relationships between variables, and any potential issues such as missing values or outliers. This process helps in formulating hypotheses, making informed decisions, and guiding subsequent analysis to extract meaningful information from the dataset.

Exploratory Data Analysis (EDA) covers a wide array of techniques and approaches, including:

- **Statistical Analysis:** Exploring data characteristics numerically to understand distributions, central tendencies, and variability.

- **Data Visualization Techniques:** Visualizing data to glean insights, discover patterns, and identify trends through plots, graphs, and charts.

- **Pattern Recognition:** Unveiling hidden structures and trends within the data, often utilizing techniques like clustering, dimensionality reduction, and anomaly detection.

- **Statistical Testing:** Using various statistical tests to validate hypotheses, compare groups, and make inferences about the data.

- **Correlation Analysis:** Assessing relationships between variables to understand how they change together or influence each other.

- **Data Cleaning and Preprocessing:** Preparing data by handling missing values, outliers, and transforming variables to make it suitable for analysis.



# Statistical Analysis

#### Descriptive Statistics

**Measures of Central Tendency:** Descriptive statistics helps to identify where the center of the data lies. This includes:

- **Mean:** The average value of the dataset.
- **Median:** The middle value when the data is arranged in ascending or descending order.
- **Mode:** The most frequently occurring value(s) in the dataset.

**Measures of Dispersion:** Descriptive statistics also explores how spread out the data points are:

- **Range:** The difference between the maximum and minimum values.
- **Variance:** A measure of how far the data points are from the mean.
- **Standard Deviation:** A measure of the average distance of data points from the mean.

**Measures of Distribution Shape:** Descriptive statistics helps to understand the shape of the data's distribution:

- **Skewness:** Indicates the asymmetry of the distribution.
- **Kurtosis:** Measures the "peakedness" or flatness of the distribution.


```python
import numpy as np

data = [12,15, 15, 18, 20, 22, 25, 30, 32, 35, 40]
```

## Mean

```python
mean_value = np.mean(data)
mean_value
```

## Median

```python
median_value = np.median(data)
median_value
```

## Mode

```python
from scipy import stats

mode_value = stats.mode(data)
mode_value.mode
```

## Range

```python
range_value = np.ptp(data)
range_value
```

## Variance

```python
variance_value = np.var(data)
variance_value
```

## Standard Deviation

```python
std_dev_value = np.std(data)
std_dev_value
```

We can obtain all of those measures by using `describe(include="all")` when working with a dataset.

# Statistical Testing

## Hypothesis testing

Hypothesis testing is a statistical method used to assess the validity of a hypothesis about a population parameter. Several common tests are:

- **T-tests:** These tests compare the means of two groups to determine if there is a significant difference between them. For instance, a t-test can help determine if the means of two samples are significantly different from each other.

- **ANOVA (Analysis of Variance):** ANOVA is used to compare means of three or more groups to see if there is a significant difference between them. It assesses whether there are statistically significant differences among the means of groups.

- **Chi-Square Tests:** Chi-square tests are used to examine the association between categorical variables. It assesses whether the observed categorical distribution differs significantly from the expected distribution.

These tests help in making statistical inferences about the population based on sample data and are widely used in various fields for hypothesis validation and decision-making.


# **Correlation Analysis:**

Correlation analysis is a statistical technique used to measure and describe the strength and direction of a relationship between two variables.

- **Pearson's correlation coefficient:** 
  - Pearson's correlation coefficient measures the linear relationship between two continuous variables. It provides a value between -1 and 1, where:
    - 1 indicates a perfect positive linear relationship,
    - -1 indicates a perfect negative linear relationship,
    - 0 indicates no linear relationship between the variables.

- **Spearman's rank correlation coefficient:** 
  - Spearman's rank correlation coefficient assesses the strength and direction of association between two ranked (ordinal) variables. It measures the monotonic relationship between variables, disregarding specific values and focusing on the ranks or orders of the data points.


# **Pattern Recognition Techniques:**
   - **Clustering algorithms (K-Means, Hierarchical Clustering):**
     - **K-Means:** A partitioning method that divides data into K clusters based on similarity.
     - **Hierarchical Clustering:** Builds a hierarchy of clusters by either merging or dividing them successively.

   - **Dimensionality reduction (PCA, t-SNE):**
     - **Principal Component Analysis (PCA):** Reduces the dimensionality of data by transforming it into a new set of variables (principal components) while retaining as much variance as possible.
     - **t-Distributed Stochastic Neighbor Embedding (t-SNE):** Reduces high-dimensional data into a lower-dimensional space, often used for visualizing high-dimensional datasets.

   - **Time series analysis (Seasonal Decomposition):**
     - **Seasonal Decomposition:** A technique to break down a time series into components like trend, seasonality, and noise to understand underlying patterns and variations.

   - **Association rule mining (Apriori Algorithm):**
     - **Apriori Algorithm:** Used to discover frequent patterns or associations in datasets, particularly in transactional databases or market basket analysis.


# **Data Cleaning and Preprocessing:**

Data cleaning and preprocessing involve several key steps:

- **Handling Missing Values:** 
  - Employ strategies like imputation (replacing missing values with substitutes) or deletion of incomplete records to manage missing data effectively.

- **Outlier Detection and Treatment:** 
  - Identify and manage outliers, which are data points significantly different from other observations in a dataset, to ensure they dont skew analysis results.

- **Feature Scaling/Normalization:** 
  - Rescale features to a similar scale or distribution, preventing any variable from dominating due to its magnitude.
  
- **Handling Categorical Variables:** 
  - Encode categorical variables appropriately to make them usable for analysis or modeling purposes.


## Tools to better understand your dataset

- **Understanding Data:**
  - **Data Types:** Use `df.dtypes` to identify variable types (numerical, categorical, etc.), crucial for further analysis.
  - **Data Size and Shape:** Determine data size and structure with `df.shape`. For extensive datasets, consider specialized tools for exploratory data analysis (EDA).

- **Handling Missing Data:**
  - **Identifying Missing Values:** Utilize `df.isnull().sum()` to pinpoint missing values. This insight aids in assessing column reliability for analysis and guides potential imputation strategies.

- **Data Summarization:**
  - **Descriptive Statistics:** Employ `df.describe()` for an overview encompassing statistical metrics like mean, median, and standard deviation.
  - **Categorical Distribution:** Use `df['column_name'].value_counts()` to reveal the frequency distribution of individual categories within a specific column.

- **Further Data Exploration:**
  - **Cross-tabulation and Pivot Tables:** Utilize functions like `pd.crosstab()` or pivot tables in Pandas to explore relationships between variables, especially useful for categorical data analysis.
  - **Handling Duplicates:** Identify and handle duplicate rows or entries in datasets using methods like `df.duplicated()` and `df.drop_duplicates()` in Pandas.
