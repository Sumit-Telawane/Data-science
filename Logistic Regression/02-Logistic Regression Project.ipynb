```python
sns.set_style("whitegrid")
sns.histplot(df['Age'], edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

# Set x-axis and y-axis limits to start from 10
plt.xlim(10,70)
plt.ylim(10, None)  # You can adjust the upper limit of the y-axis as needed

plt.show()
```

**Create a jointplot showing the kde distributions of Daily Time spent on site vs. Age.**

```python
from scipy.stats import pearsonr
joint_plot = sns.jointplot(x='Age', y='Daily Time Spent on Site', data=df, kind='kde', color='Red',fill=True)

pearson_corr, p_value = pearsonr(df['Age'], df['Daily Time Spent on Site'])

joint_plot.ax_joint.annotate(
    f"Pearsonr = {pearson_corr:.2f}, p-value = {p_value:.4f}",
    xy=(0.8, 0.98),
    xycoords='axes fraction',
    ha='center',
    fontsize=8,
    color='Black'
)

plt.show()
```

** Create a jointplot of 'Daily Time Spent on Site' vs. 'Daily Internet Usage'**

```python
joint_plot = sns.jointplot(x='Daily Time Spent on Site', y='Daily Internet Usage', data=df, kind='scatter', edgecolor='w', color='skyblue')

pearson_corr, p_value = pearsonr(df['Daily Time Spent on Site'], df['Daily Internet Usage'])

joint_plot.ax_joint.annotate(
    f"Pearsonr = {pearson_corr:.2f}, p-value = {p_value:.4f}",
    xy=(0.8, 0.98),
    xycoords='axes fraction',
    ha='center',
    fontsize=8,
    color='Black'
)

plt.show()
```

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split

# Features (X)
X = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]

# Target variable (y)
y = df['Clicked on Ad']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
logistic_model = LogisticRegression(random_state=42)

# Train the model on the training set
logistic_model.fit(X_train, y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
y_pred = logistic_model.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report

classification_rep = classification_report(y_test, y_pred)

# Print the classification report
print(classification_rep)
```
